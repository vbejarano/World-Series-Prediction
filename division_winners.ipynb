```python
# Read Data File and store into Pandas Data Frames
team_batting = pd.read_csv(team_batting_csv)
team_fielding = pd.read_csv(team_fielding_csv)
team_pitching = pd.read_csv(team_pitching_csv)
team_waa = pd.read_csv(team_waa_csv)
ari = pd.read_csv(ari_csv)
atl = pd.read_csv(atl_csv)
bal = pd.read_csv(bal_csv)
bos = pd.read_csv(bos_csv)
chc = pd.read_csv(chc_csv)
chw = pd.read_csv(chw_csv)
cin = pd.read_csv(cin_csv)
cle = pd.read_csv(cle_csv)
col = pd.read_csv(col_csv)
det = pd.read_csv(det_csv)
hou = pd.read_csv(hou_csv)
kcr = pd.read_csv(kcr_csv)
laa = pd.read_csv(laa_csv)
lad = pd.read_csv(lad_csv)
mia = pd.read_csv(mia_csv)
mil = pd.read_csv(mil_csv)
minn = pd.read_csv(min_csv)
nym = pd.read_csv(nym_csv)
nyy = pd.read_csv(nyy_csv)
oak = pd.read_csv(oak_csv)
phi = pd.read_csv(phi_csv)
pit = pd.read_csv(pit_csv)
sdp = pd.read_csv(sdp_csv)
sea = pd.read_csv(sea_csv)
sfg = pd.read_csv(sfg_csv)
stl = pd.read_csv(stl_csv)
tbr = pd.read_csv(tbr_csv)
tex = pd.read_csv(tex_csv)
tor = pd.read_csv(tor_csv)
wsn = pd.read_csv(wsn_csv)
```

```python
#concat to put all teams in one dataframe
team_wl = pd.concat([ari, atl, bal, bos, chc, chw, cin, cle, col, det, hou, kcr, laa, lad, mia, mil, minn, nym, nyy,
                   oak, phi, pit, sdp, sea, sfg, stl, tbr, tex, tor, wsn])
```

```python
# view team_wl
team_wl.head()
```

```python
# merge batting and fielding statistics
combined_stats1 = pd.merge(team_batting, team_fielding,  on = "Tm", how = "inner")
combined_stats1.head()
```

```python
# add pitching statistics
combined_stats =pd.merge(combined_stats1, team_pitching,  on = "Tm", how = "inner")
combined_stats.head()
```

```python
# add hitting, fielding and pitching statistics to the team's w/l records
team_combined_stats = pd.merge(team_wl, combined_stats, on = "Tm", how = "inner")
team_combined_stats.head()
```

```python
# dropping summary lines from table where no team is listed. Each team's file contained sum/summary totals at the bottom
updated_team_stats = team_combined_stats.drop(team_combined_stats.index[20:50])
updated_team_stats.head()
```

```python
#dropping uneccessary columns
team_stats_drop = updated_team_stats.drop(columns =["#Bat", "BatAge", "G_x", "PA"])
team_stats_drop.head()
```

```python
team_stats_drop.to_csv("Data/updated_team_stats.csv")
```

```python
# separate w/l to use to train
yTr= team_stats_drop['W/L']
xTr = team_stats_drop.drop(columns =['W/L','Tm', 'Opp','R','RA','Gm#'])
xTr.head()
```

```python
# stats included wo with either the W or L indicator for walk off wins/losses. dropping the wo from both outcomes. setting y
y1 = yTr.replace('L-wo', 'L')
y = y1.replace('W-wo','W')
y.head()
```

```python
# converting the W/L column to a numeric value for training purposes. Dropping the L column to leave 1 = W, 0 = L
y2 =pd.get_dummies(y)
y2 =y2.drop(columns =['L'], axis = 1)
y2.head()
```

```python
# defining X
X = xTr
X.head()
```

```python
# setting feature names and test/train split 
feature_names = X.columns
X_train, X_test, y_train, y_test = train_test_split(X, y2, random_state=1, stratify=y)
```

```python
# fitting model
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
```

```python
# getting score for training data
rf.score(X_train, y_train)
```

```python
# getting score for test data
rf.score(X_test, y_test)
```

```python
# getting weight for each feature
sorted(zip(rf.feature_importances_, feature_names), reverse = True)
```
